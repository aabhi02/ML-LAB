```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('ENJOYSPORT.csv')
```

```python
print(data)
```

```python
data_array=np.array(data)[:,:-1]
```

```python
data_label=np.array(data['EnjoySport'])
```

```python
#find s

ans=[0, 0, 0, 0, 0, 0]
def find_s(data, label):
    for i in range(len(label)):
        if label[i]==1:
            for j in range(len(ans)):
                    ans[j]=data[i][j]
            break
    for i in range(len(data)):
        if label[i]==1:
            for j in range(len(ans)):
                if ans[j]!=data[i][j]:
                    ans[j]='?'
            
```

```python
find_s(data_array, data_label)
```
